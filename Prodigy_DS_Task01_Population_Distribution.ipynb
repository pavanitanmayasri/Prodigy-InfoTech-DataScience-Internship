```python
df = pd.read_csv(
    'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv',
    skiprows=4
)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
population_2022 = df['2022'].dropna()
```

```python
plt.figure(figsize=(10,6))
plt.hist(population_2022, bins=30)
plt.title('Distribution of World Population by Country (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

```python
top10 = df[['Country Name', '2022']].dropna().sort_values(
    by='2022', ascending=False
).head(10)

plt.figure(figsize=(12,6))
plt.barh(top10['Country Name'], top10['2022'])
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
